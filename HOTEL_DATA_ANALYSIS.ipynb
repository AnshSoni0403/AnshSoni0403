```python
cancelled_perc = df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)
```

This shows that 62% is not cancelled and 37% is cancelled and hence it is a large amount.

```python
pl.figure(figsize = (5,5))
pl.title("Reservation status count")
pl.bar(['Not Cancelled','Cancelled'],df['is_canceled'].value_counts(),width = 0.4,color='r')
pl.show()
```

```python
pl.figure(figsize = (8,4))
ax1 = sns.countplot(x = 'hotel',hue = 'is_canceled',data = df,palette = 'rocket')
legend_labels = ['Not Cancelled','Cancelled']
pl.title('Reservation status in different hotels')
pl.xlabel('Hotel')
pl.ylabel('No of reservations')
```

Cancellation in City Hostel is more than Resort Hotel.

```python
cancelled_per_hotel = df.groupby('hotel')['is_canceled'].value_counts(normalize=True)
print(cancelled_per_hotel)
```

```python
resort_hotel = df['is_canceled'].value_counts(normalize=True)
print(resort_hotel)
```

```python
city_hotel = df['is_canceled'].value_counts(normalize=True)
print(city_hotel)
```

```python
#adr average daily rate
resort_hotel_adr = df[df['hotel'] == 'Resort Hotel'].groupby('reservation_status_date')[['adr']].mean()
city_hotel_adr = df[df['hotel'] == 'City Hotel'].groupby('reservation_status_date')[['adr']].mean()

print(resort_hotel_adr)
print(city_hotel_adr)
```

```python
# Plot the average daily rates
pl.figure(figsize = (20,8))
pl.title("Average daily rate in each hotel")

# Use the correct variables for plotting
pl.plot(resort_hotel_adr.index,resort_hotel_adr['adr'],label="Resort Hotel")
pl.plot(city_hotel_adr.index,city_hotel_adr['adr'],label="City Hotel")

pl.xlabel("Reservation status date")
pl.ylabel("Average daily rate")
pl.legend(fontsize=20)
pl.show()
```

```python
df['month'] = pd.to_datetime(df['reservation_status_date']).dt.month

ax1 = sns.countplot(x = 'month',hue = 'is_canceled',data = df,palette = 'bright')
pl.title('Reservation status in each month')
pl.xlabel('Month')
pl.ylabel('No of reservations')

```

Most cancellation in January.
least in August


```python
pl.figure(figsize=(15, 8))
pl.title("Reservation prices in each month (for cancelled)")
# Pass the DataFrame to the 'data' argument, and specify x and y using column names
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(),palette="rocket")
pl.xlabel("Month")
pl.ylabel("Total revenue")
pl.show()
```

Since prices were hired - so at last moment reservations were cancelld

```python
top_10_country = df[df['is_canceled'] == 1]['country'].value_counts().head(10)
print(top_10_country)
pl.figure(figsize = (8,8))
pl.title("Cancellation in each country for top 10 country")
pl.pie(top_10_country,autopct = '%1.1f%%', labels = top_10_country.index)
pl.show()
```

highest cancellation in portugese
